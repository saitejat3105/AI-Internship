```python
#covert object data to normal data
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
df.buying.value_counts()
```

```python
df.buying=le.fit_transform(df.buying)
df.maint=le.fit_transform(df.maint)
df.lug_boot=le.fit_transform(df.lug_boot)
df.safety=le.fit_transform(df.safety)
#normally there is no need to transform outcomr tragrt column but it seems a update to objrct datatypr
#lets got normally only new update
```

```python
df.head()
```

```python
X=df.iloc[:,:-1]
X.head()
```

```python
y=df.outcome
y.head()
```

```python
#first check withoutbbalanceing the accuiracy
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=101)
```

```python
X_train.shape
```

```python
#for multiclasses knn works best
#it is based on eculidian distance based alg
#for eg we have 1000 observations
#first it selects random data from 1000 obs and calls centroid
#checks distance b/w each points we randomly selects 2 data points
#how 2 means we have a parameter k if k=2 2 points selects and checks each and every datapoint dist b/w with these 2 points
#if k=3 then 3 points and check dist
#after we split by malking lines like in graph(assume many points in graph)
#which is near to centroid will become part/class of the 2 or 3 points
#like red class bluee class eg this is how it clusters the data points
#it claculates distance on eculidhnace dist alg
```

```python
from sklearn.neighbors import KNeighborsClassifier
model=KNeighborsClassifier()
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
accuracy_score(y_test,y_pred)
```

```python
pd.crosstab(y_test,y_pred)
```

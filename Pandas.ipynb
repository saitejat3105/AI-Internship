#merge

```python
#2 data set merge
dd=pd.DataFrame({'name':['sai','teja','gate','isagi'],'english':[78,89,99,90]})
dd
```

```python
dm=pd.DataFrame({'name':['sai','teja','gate','isagi'],'maths':[23,45,67,77]})
dm
```

```python
m1=pd.merge(dd,dm, on ='name')# only 2 possible not 3
m1
```

```python
dh=pd.DataFrame({'name':['sai','teja','gate','isagi'],'python':[23,45,77,88]})
dh
```

```python
m2=pd.merge(m1,dh, on ='name')
m2
```

#pie chart

```python
data=[2,4,6,8]
plt.pie(data)
```

```python
plt.pie(data) # only diag
plt.show
```

```python
plt.pie(data,[0,0,0,0.2])#explode
```

```python
plt.pie(data,[0,0.9,0,0])
```

```python
plt.pie(data,[0,0,.09,0],labels=['t1','t2','t3','t4'])
```

```python
plt.pie(data,[0,0,.09,0],labels=['t1','t2','t3','t4'],autopct="%1.1f%%")#autopct percentages
```

```python
plt.pie(data,[0,0,.09,0],labels=['t1','t2','t3','t4'],autopct="%1.1f%%",shadow=True) #shadows
```

```python
plt.pie(data,[0,0,.09,0],labels=['t1','t2','t3','t4'],autopct="%1.1f%%",shadow=True,colors=['yellow','pink','green','blue'])
plt.title(["pie "]) #title
plt.show()
```

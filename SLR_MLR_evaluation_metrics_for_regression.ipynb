```python
#R2 Score
#coeff of determination or R^2 is another metric used for evaluating performance of regression model
#this metric helps us to compare our current model with a const baseline and tells us how much our model is better
#R^2 is scale free score that implies it doesnt matter whether the values are too large or too small
#R^2 will always be less than or equal to 1
#R^2=1-(mse(model)/mse(baseline))           mse mean square error error ,minimumn
```

```python
#SSresiduals=SUM(y-y^)^2
#SStotal=SUM(y-yavg)^2
#  R^2=1-(SSres/SStot)  res resuduals error
```

```python
#metrics library
```

```python
#mean square error
#most prefered for regression
#it is simply avg of squared diff btw target value and predicted value by regression model
#as it squares the diff it penalizes even a small error which leads to over estimation of how bad the model is
#it is preffered more than other metrics cause it is differentiable and hence can be optimized better
#MSE=1/n sum(y-y^)^2  the square of diff btw actual and predicted
```

```python
#root mean square error: square root of mean suqare error
#RMSE is not most widely used metric for regression tasks and is the square root of the averaged squared diff btw
#target value and value predicted by model
#it is preffered more in some cases cause the errors are 1st squared before averaging which posess high penality on
# large errors
#RMSE=sqrt(summa(y-y^)^2/N) less rmse best model
```
